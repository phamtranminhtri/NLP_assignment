```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
metrics = ["loss", "acc", "precision", "recall", "f1"]
```

```python
def plot_train_val(log_dir, metric):
    train_file = log_dir + "/" + "train_" + metric + ".csv"
    val_file = log_dir + "/" + "val_" + metric + ".csv"

    train_df = pd.read_csv(train_file)
    val_df = pd.read_csv(val_file)

    plt.figure()

    plt.plot(train_df["Step"], train_df["Value"], label=f"Train {metric}")
    plt.plot(val_df["Step"], val_df["Value"], label=f"Validation {metric}")

    plt.xlabel("Step")
    plt.ylabel("Loss")
    plt.title(f"Train & Validation {metric}")
    plt.legend()
    plt.grid(True)

    plt.show()
```

```python
print("BERT")

for metric in metrics:
    plot_train_val("bert/csv_logs", metric)
```

```python
print("SBERT")

for metric in metrics:
    plot_train_val("sbert/csv_logs", metric)
```
